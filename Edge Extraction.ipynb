```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Read the image
def util(img_dir):
  img = cv2.imread(img_dir, cv2.IMREAD_GRAYSCALE)

  # Apply the Sobel operator to the x and y directions
  sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
  sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

  # Calculate the gradient magnitude
  grad_magnitude = np.sqrt(np.square(sobelx) + np.square(sobely))

  # Normalize the gradient magnitude to the range [0, 255]
  grad_magnitude = (grad_magnitude / grad_magnitude.max()) * 255

  # Convert the gradient magnitude to an 8-bit integer
  grad_magnitude = np.uint8(grad_magnitude)
  temp = img_dir[45:].split('/')[-1]
  temp_len = len(img_dir) - len(temp)
  save_dir = '/content/drive/MyDrive/Capstone_Project/Edges_Data/' + img_dir[45:temp_len]
  # print(save_dir)
  os.chdir(save_dir)
  # Show the edge image
  filename = img_dir[45:].split('/')[-1]
  # print(filename)
  cv2.imwrite(filename, grad_magnitude)
  cv2_imshow(grad_magnitude)
```
